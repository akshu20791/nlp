```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = [
    "watch",
    "watching",
    "watched",
    "watches"
]

for word in words:
    print(
        word,
        " ---> ",
        stemmer.stem(word)
    )
```
